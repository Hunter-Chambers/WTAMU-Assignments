## Perl

- History 
    - Designed by Larry Wall—first released in 1987
    - General-purpose Unix scripting language to make report processing easier
    - "the Swiss Army chainsaw of scripting languages"
- Characteristics
    - Always compiled
    - Variables are statically typed but implicitly declared
    - Three distinctive namespaces, denoted by the first character of a variable's name
    - Powerful, but somewhat dangerous
- Uses    
    - Gained widespread use for CGI programming on the Web
    - Also used for a replacement for UNIX system administration language
- https://perldoc.perl.org/
- https://www.geeksforgeeks.org/hello-world-program-in-perl/

## JavaScript

- History
    - Created by Brendan Eich in 1995 during his time at Netscape Communications
    - Began at Netscape, but later became a joint venture of Netscape and Sun Microsystems
        - Initially programming language for Netscape web browser
        - Need for computation accociated with completely static HTML pages
- Characteristics 
    - Embedded in HTML documents
    - Purely interpreted
    - Dynamic nature of strings and arrays
    - Dynamic typing
    - Related to Java only through similar syntax
- Uses
    - A client-side HTML-embedded scripting language
    - Often used to create dynamic HTML documents
- https://developer.mozilla.org/en-US/docs/Web/JavaScript
- https://javascript.info/hello-world

##  PHP

- History
    - Designed by Rasmus Lerdorf in 1994
    - Common Gateway Interface (CGI) in C
        - To maintain personal homepage - track traffic
        - Personal Home Page/Forms Interpreter - PHP/FI
- Characteristics
    - Similar to Javascript by
        - Dynamic nature of strings and arrays
        - Dynamic typing
        - Syntax
    - Purely interpreted
    - Object Oriented support added from second release
        - Does not support
            - Abstract classes or interfaces
            - Destructors or access control for class members
- Uses 
    - A server-side HTML-embedded scripting language
    - often used for form processing and database access through the Web

- https://www.php.net/docs.php
- https://www.youtube.com/watch?v=xMRjJwPSsWQ

##  Ruby

- History
    - Designed in Japan by Yukihiro Matsumoto (a.k.a, “Matz”) in the 1990's
    - First release in 1996
    - Began as a replacement for Perl and Python
        - Programming should be fun for programmers
        - Software should be understood by humans first
    - First programming language developed in Japan
- Characteristics
    - A pure object-oriented scripting language
         - All data are objects
    - Most operators are implemented as methods, which can be redefined by user code
    - Purely interpreted
- Uses
    - General purpose scripting language
    - Mainly used for web applications and system utilities
- https://www.ruby-lang.org/en/documentation/
- https://en.wikibooks.org/wiki/Ruby_Programming/Hello_world

## Lua

- History
    - Originally designed in 1993 in Brazil 
        -  Roberto Ierusalimschy, Luiz Henrique de Figueiredo, and Waldemar Celes
    - Language for extending software applications
- Characteristics
    - Multi-paradigm language
        - An Object Oriented interpreted scripting language
        - Supports closures - could be used for functional programming
    - Type checked but dynamically typed
    - Does not contain explicit support for inheritance
        - Allows it to be implemented with metatables
    - Supports lists, tuples, and hashes, all with its single data structure, the table
    - Easily extendable
- Uses
    - Game programming :) 
    - Web applications
        - CGI programming and form processing
    - Image processing
- https://www.lua.org/docs.html
- https://www.lua.org/pil/1.html

## Python

- History
    - Its initial design was in 1991
        - by Guido van Rossum at Stichting Mathematisch Centrum in the Netherlands
    - Developed for emphasis on code readibility 
    - Syntax allows programmers more expressivity in fewer lines of code
- Characteristics
    - Easy to install
    - Easy to use
    - Supports different programming approaches
        - An Object Oriented interpreted scripting language
    - Type checked but dynamically typed
    - Supports lists, tuples, and hashes
- Uses
    - According to IEEE survey Python is the number one programming language
    - Python is a major language in majority of big companies
        - Google, Spotify, Netflix, Dropbox, HPE, Intel, Salesforce...
    - Working on some of the most exciting technologies today
        - Machine learning
        - Artificial Intelligence (AI)
        - Big Data 
        - Robotics
        - Web API
        - Cyber Security, one of the top software challenges of our time, is also driven by Python
    - There is a tremendous support 
        - Both for developing
        - and Learning
            - http://www.python.org
            - http://learnpythonthehardway.org/book/
            - https://www.codecademy.com/tracks/python
            - https://python-textbok.readthedocs.io/en/1.0/
        - Plenty of courses on coursera, edx and other MOOCs
    - It's now the most popular introductory language taught in universities
- https://docs.python.org/3/
- https://www.python.org/dev/peps/#introduction

## Zen of Python

Long time Pythoneer Tim Peters succinctly channels the BDFL's guiding principles for Python's design into 20 aphorisms, only 19 of which have been written down

1. Beautiful is better than ugly
2. Explicit is better than implicit
3. Simple is better than complex
4. Complex is better than complicated
5. Flat is better than nested
6. Sparse is better than dense
7. Readability counts
8. Special cases aren't special enough to break the rules
9. Although practicality beats purity
10. Errors should never pass silently
11. Unless explicitly silenced
12. In the face of ambiguity, refuse the temptation to guess
13. There should be one -- and preferably only one -- obvious way to do it
14. Although that way may not be obvious at first unless you're Dutch
15. Now is better than never
16. Although never is often better than *right* now
17. If the implementation is hard to explain, it's a bad idea
18. If the implementation is easy to explain, it may be a good idea
19. Namespaces are one honking great idea -- let's do more of those
